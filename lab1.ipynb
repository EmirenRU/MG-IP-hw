```python
def draw_plane(n):
    fig, ax = plt.subplots(figsize=(6, 6), dpi=200)
    
    ax.set_xlim([-n, n])
    ax.set_ylim([-n, n])
    
    ax.set_xlabel(r'$O_x$', fontsize=14)
    ax.set_ylabel(r'$O_y$', fontsize=14)
    
    ax.grid(True)
    
    ax.set_xticks(range(-n, n+1))
    ax.set_yticks(range(-n, n+1))
    ax.axhline(0, color='black',linewidth=1)
    ax.axvline(0, color='black',linewidth=1)

draw_plane(5)
```

```python
import matplotlib.pyplot as plt

def draw_manual_grid(n):
    fig, ax = plt.subplots(figsize=(6, 6), dpi=200)
    
    ax.set_xlim([-n, n])
    ax.set_ylim([-n, n])
    
    ax.set_xlabel(r'$O_x$', fontsize=14)
    ax.set_ylabel(r'$O_y$', fontsize=14)
    
    ax.grid(False)

    ax.set_xticks(range(-n, n+1))
    ax.set_yticks(range(-n, n+1))

    for i in range(-n, n+1):
        ax.plot([i, i], [-n, n], color='gray', linestyle='--', linewidth=0.5)  
        ax.plot([-n, n], [i, i], color='gray', linestyle='--', linewidth=0.5)  
    
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    plt.show()

draw_manual_grid(5)
```

```python
def draw_infinite_grid(n):
    fig, ax = plt.subplots(figsize=(6, 6), dpi=200)
    
    ax.set_xlim([-n, n])
    ax.set_ylim([-n, n])
    
    ax.set_xlabel(r'$O_x$', fontsize=14)
    ax.set_ylabel(r'$O_y$', fontsize=14)

    ax.set_xticks(range(-n, n+1))
    ax.set_yticks(range(-n, n+1))
    
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    for i in range(-n, n+1):
        ax.axline((i, -n), slope=0, color='gray', linestyle='--', linewidth=0.5)  # –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏
        ax.axline((-n, i), slope=float('inf'), color='gray', linestyle='--', linewidth=0.5)  # –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏

    plt.show()

draw_infinite_grid(5)
```

```python
def draw_diagonal_lines(n):
    fig, ax = plt.subplots(figsize=(6, 6), dpi=200)
    
    ax.set_xlim([-n, n])
    ax.set_ylim([-n, n])
    
    ax.set_xlabel(r'$O_x$', fontsize=14)
    ax.set_ylabel(r'$O_y$', fontsize=14)
    
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    for i in range(-n, n+1):
        ax.plot([i, i], [-n, n], color='gray', linestyle='--', linewidth=0.5)  
        ax.plot([-n, n], [i, i], color='gray', linestyle='--', linewidth=0.5)  
    
    for i in range(-n, n):
        for j in range(-n, n):
            ax.plot([i, i+1], [j, j+1], color='gray', linewidth=1)  # –î–∏–∞–≥–æ–Ω–∞–ª—å (–≤–≤–µ—Ä—Ö –≤–ø—Ä–∞–≤–æ)

    plt.show()

draw_diagonal_lines(5)
```

- –ü—Ä–∏–º–µ–Ω–∏—Ç–µ –∫ –≤–µ–∫—Ç–æ—Ä–∞–º —Ç–∞–∫–æ–µ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, —á—Ç–æ–±—ã –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —É–≥–æ–ª –ø–æ–º–µ–Ω—è–ª —Å–≤–æ–π –∑–Ω–∞–∫. –ö–∞–∫–æ–º—É —É—Å–ª–æ–≤–∏—é –¥–æ–ª–∂–Ω–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—Ç—å –º–∞—Ç—Ä–∏—Ü–∞ —Ç–∞–∫–æ–≥–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è?

(–º–∞—Ç—Ä–∏—Ü–∞ –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –º–∞—Ç—Ä–∏—Ü–µ–π –æ—Ç—Ä–∞–∂–µ–Ω–∏—è –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–∞—á–∞–ª–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç)

- –ù–∞—Ä–∏—Å—É–π—Ç–µ —Ç–æ –∂–µ —Å–∞–º–æ–µ –ø–µ—Ä–µ–∫–ª—é—á–∏–≤—à–∏—Å—å –≤ –ø–æ–ª—è—Ä–Ω—É—é —Å–∏—Å—Ç–µ–º—É –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç, –ø–æ–ª—É—á–∏–≤—à–µ–µ—Å—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –ø–æ—Ö–æ–∂–µ –Ω–∞ —Ä–∏—Å—É–Ω–∫–∏ 3 –∏ 4.

```python
x1,y1 = 2,1
x2,y2 = -1,1

A = np.array([x1, y1])
B = np.array([x2, y2])

theta_A = np.arctan2(A[1], A[0])
theta_B = np.arctan2(B[1], B[0])

delta_theta = theta_B - theta_A

plt.figure(figsize=(6,6), dpi=200)
plt.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='r', label='Transformed Vector A')
plt.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='b', label='Transformed Vector B')

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.grid(True)
plt.legend()
plt.title('Transformed Vectors A and B')
```

```python
# fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
fig = plt.figure(num=10, figsize=(3,3), dpi=200)
ax = fig.add_subplot(1, 1, 1, projection='polar')
r_A = np.linalg.norm(A)
r_B = np.linalg.norm(B)
plt.yticks(np.arange(0, np.ceil(np.linalg.norm(A)), step=1))

plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.arrow(0,0,theta_A, r_A,head_width=0.07, color='blue')
plt.arrow(0,0,theta_B, r_A,head_width=0.07, color='green')
ax.legend()
ax.set_title(r'Vectors in Polar Coordinates')
```

```python
A_transformed = -A
B_transformed = -B

plt.figure(figsize=(6,6), dpi=200)
plt.quiver(0, 0, A_transformed[0], A_transformed[1], angles='xy', scale_units='xy', scale=1, color='r', label='Transformed Vector A')
plt.quiver(0, 0, B_transformed[0], B_transformed[1], angles='xy', scale_units='xy', scale=1, color='b', label='Transformed Vector B')

plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Transformed Vectors A and B')

theta_A_transformed = np.arctan2(A_transformed[1], A_transformed[0])
theta_B_transformed = np.arctan2(B_transformed[1], B_transformed[0])

delta_theta_transformed = theta_B_transformed - theta_A_transformed

print(f'Transformed oriented angle between vectors: {np.degrees(delta_theta_transformed)} degrees')

plt.show()
```

- –í—ã–¥–µ–ª–∏—Ç–µ –≤–µ—Ä—à–∏–Ω—ã –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ –º–∞—Ä–∫–µ—Ä–Ω—ã–º–∏ —Ç–æ—á–∫–∞–º–∏. –ö–∞–∫ –Ω–∞—Å—Ç—Ä–æ–∏—Ç—å —Ä–∞–∑–º–µ—Ä –º–∞—Ä–∫–µ—Ä–æ–≤? –¶–≤–µ—Ç? –§–æ—Ä–º—É?

- –ü–æ–∫–∞–∂–∏—Ç–µ —Å—Ç—Ä–µ–ª–∫–∞–º–∏ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –æ–±—Ö–æ–¥–∞ –≤–µ—Ä—à–∏–Ω, –∫–∞–∫ —ç—Ç–æ –ø–æ–∫–∞–∑–∞–Ω–æ –Ω–∞ —Ä–∏—Å—É–Ω–∫–∞—Ö 5 –∏ 6

- —Ä–∏—Å. 5 –¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –ø–ª–æ—â–∞–¥—å—é (P1{1;3} P2{1,1}, P{4,1})
- —Ä–∏—Å. 6 –¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –ø–ª–æ—â–∞–¥—å—é (P1{-3;-1} P2{-1,-1}, P{-1,-4})

```python
n = 3
low = -5
high = 5

vertices = np.random.randint(low,high, (n,2))
print(vertices.shape[0])
# vertices = np.array([
#     [0,0],
#     [0,2],
#     [2,2],
#     [2,0]
# ])

# –§–æ—Ä–º—É–ª–∞ –ø–ª–æ—â–∞–¥–∏ –ì–∞—É—Å—Å–∞
def oriented_area(vertices):
    area = 0
    for i in range(n):
        x1,y1 = vertices[i]
        x2,y2 = vertices[(i+1) % n]
        area += (x1 * y2 - x2 * y1)
    return 0.5 * area

def draw_polygons(vertices):
    fig, ax = plt.subplots(figsize=(6, 6), dpi=200)
    poly = patches.Polygon(xy=vertices, closed=True, fill=False, edgecolor='b', linewidth=2)
    ax.add_patch(poly)
    ax.set_xlim([low*2, high*2])
    ax.set_ylim([low, high])
    
    ax.set_xlabel(r'$O_x$', fontsize=14)
    ax.set_ylabel(r'$O_y$', fontsize=14)
    
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    plt.grid(True)
    x_coords = vertices[:, 0]
    y_coords = vertices[:, 1]

    
    ax.plot(x_coords, y_coords, 'ro', markersize=8)

    for i in range(3):
        x1,y1 = vertices[i]
        x2,y2 = vertices[(i+1) % n]
        plt.arrow(x1,y1,x2-x1,y2-y1,head_width=0.3)
    
    # for i in range(low, high+1):
        # ax.plot([i, i], [-n, n], color='gray', linestyle='--', linewidth=0.5)  
        # ax.plot([-n, n], [i, i], color='gray', linestyle='--', linewidth=0.5)  
    plt.show()

draw_polygons(vertices)
x_coords = -vertices[:, 0]
y_coords = -vertices[:, 1]
negativeVert = np.column_stack([x_coords, y_coords])

 
draw_polygons(negativeVert)

print(f' {type(vertices)} {vertices}  {oriented_area(vertices)}')
print(f' {type(negativeVert)} {negativeVert}  {oriented_area(negativeVert)}')
```

5. –ù–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Ä–∏—Å—É–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ —Å —Ü–µ–Ω—Ç—Ä–æ–º –≤ –∑–∞–¥–∞–Ω–Ω–æ–π —Ç–æ—á–∫–µ, –∞ —Ç–∞–∫–∂–µ –≤–ø–∏—Å–∞–Ω–Ω—É—é –∏ –æ–ø–∏—Å–∞–Ω–Ω—É—é –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ —ç—Ç–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ —Å–º. —Ä–∏—Å. 8. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–æ—Ä–æ–Ω –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ ùëõ-—É–≥–æ–ª—å–Ω–∏–∫–∞ –∏ –¥–ª–∏–Ω—É –µ–≥–æ —Å—Ç–æ—Ä–æ–Ω—ã –Ω—É–∂–Ω–æ –∑–∞–¥–∞—Ç—å –≤ –≤–∏–¥–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞. –í–º–µ—Å—Ç–æ –¥–ª–∏–Ω—ã —Å—Ç–æ—Ä–æ–Ω—ã –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ä–∞–¥–∏—É—Å ùëü –≤–ø–∏—Å–∞–Ω–Ω–æ–π –∏–ª–∏ —Ä–∞–¥–∏—É—Å ùëÖ –æ–ø–∏—Å–∞–Ω–Ω–æ–π –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–µ–π.

- –í—Å–µ –≤–µ—Ä—à–∏–Ω—ã –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –≤—ã–¥–µ–ª–µ–Ω—ã –º–∞—Ä–∫–µ—Ä–∞–º–∏ –∏ –ø–æ–¥–ø–∏—Å–∞–Ω—ã –∫–∞–∫ ùëÉ1, ùëÉ2 –∏ —Ç.–¥. –ù—É–º–µ—Ä–∞—Ü–∏—è —Ç–æ—á–µ–∫ –¥–æ–ª–∂–Ω–∞ –æ—Å—É—â–µ—Å—Ç–≤–ª—è—Ç—å—Å—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏, –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç ùëõ

- –†–∞–¥–∏—É—Å—ã –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–µ–π –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –Ω–∞—Ä–∏—Å–æ–≤–∞–Ω—ã –≤ –≤–∏–¥–µ –æ—Ç—Ä–µ–∑–∫–æ–≤.

- –°—Ç–æ—Ä–æ–Ω—ã –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –Ω–∞—Ä–∏—Å–æ–≤–∞–Ω—ã –±–æ–ª–µ–µ —Ç–æ–ª—Å—Ç—ã–º–∏ –ª–∏–Ω–∏—è–º–∏, —á–µ–º –æ—Å—Ç–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è.

- –û–ø–∏—Å–∞–Ω–Ω–∞—è –∏ –≤–ø–∏—Å–∞–Ω–Ω–∞—è –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –Ω–∞—Ä–∏—Å–æ–≤–∞–Ω—ã —Ä–∞–∑–Ω—ã–º–∏ —Ü–≤–µ—Ç–∞–º–∏.

- –ú–∞—Å—à—Ç–∞–± –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –Ω–µ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –∏—Å–∫–∞–∂–µ–Ω (–æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –Ω–µ –¥–æ–ª–∂–Ω—ã –≤—ã–≥–ª—è–¥–µ—Ç—å –∫–∞–∫ —ç–ª–ª–∏–ø—Å—ã).

- –ö–∞—Ä—Ç–∏–Ω–∫–∞ –¥–æ–ª–∂–Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –ø–µ—Ä–µ—Ä–∏—Å–æ–≤—ã–≤–∞—Ç—å—Å—è –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ ùëõ.

```python
# defined
n = 12
a = 3
# found in action
alpha = 360 / n
theta = alpha / 2
theta_rad = theta * (np.pi / 180)
r =  (a*0.5) / np.tan(theta_rad)
R = r / np.cos(theta_rad)

angles = np.linspace(0,2*np.pi, n, endpoint=False)
x_coords = R * np.cos(angles)
y_coords = R * np.sin(angles)

fig, ax = plt.subplots(figsize=(6, 6), dpi=200)

circle_outside = plt.Circle((0, 0), R, color='blue', fill=False, linestyle='dashed', linewidth=2)
ax.add_patch(circle_outside)
circle_inside = plt.Circle((0, 0), r, color='red', fill=False, linestyle='dashed', linewidth=2)
ax.add_patch(circle_inside)

for i in range(n):
    # R
    ax.plot([0, x_coords[i]], [0, y_coords[i]], color='green', linestyle='dotted', linewidth=2)

    # A
    plt.plot([x_coords[i], x_coords[(i + 1) % n]], [y_coords[i], y_coords[(i + 1) % n]], color='black', linewidth=3)

    plt.plot([(x_coords[i] + x_coords[(i + 1) % n]) / 2, 0], [(y_coords[i] + y_coords[(i + 1) % n]) / 2, 0], color='orange', linewidth=0.75, linestyle='--')
    
    
    # Markers
    ax.plot(x_coords[i], y_coords[i], marker='o', color='black') 
    ax.text(x_coords[i] * 1.1, y_coords[i] * 1.1, f'P{i+1}', fontsize=12, ha='center', va='center')

    

plt.xlim(-R - 1, R + 1)
plt.ylim(-R - 1, R + 1)
ax.set_aspect('equal', adjustable='box')  
plt.grid(True) 

plt.title(f'–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π {n}-—É–≥–æ–ª—å–Ω–∏–∫ –∏ –µ–≥–æ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏')
plt.xlabel(r'$O_x$') 
plt.ylabel(r'$O_y$') 
plt.show() 
```
